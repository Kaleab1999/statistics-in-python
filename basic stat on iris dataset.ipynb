importting necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

loading the iris dataset

```python
iris=pd.read_csv('Iris.csv')
iris.head()
```

```python
#checking missing values
col=0
for i in iris.columns:
    if(iris[i].isnull().sum()>0):
        co+=1
        m=iris[i].isnull().sum()/iris.shape[0]
        print('{} has {}'.format(i,m))
print('{} features have missing values'.format(col))
```

```python
# to calculate mean,median, mode,q1,q3,iqr
numerical=iris.select_dtypes(exclude=['object']).columns.tolist()
for i in numerical:
    print('some basic stat for {} column'.format(i))
    me=iris[i].median()
    m=iris[i].mean()
    #mo=iris[i].mode()
    q1=np.percentile(iris[i],25)
    q3=np.percentile(iris[i],75)
    st=np.std(iris[i])
    iqr=q3-q1
    va=pow(st,2)

    print('*median {}'.format(me))
    print('*mean {}'.format(m))
    #print('*mode {}'.format(mo))
    print('*std  {}'.format(st))
    print('*variance  {}'.format(va))
    print('*1st quartile {} and 3rd quartile {}'.format(q1,q3))
    print('*inter quartile range {}'.format(iqr))
```

plotting the measure of central tendency towards the data

```python
for i in numerical:
    mea=iris[i].mean()
    #st=df[i].std()
    median=iris[i].median()
    df2 = pd.DataFrame({
      'data': iris[i],
      'mean': [mea for i in range(len(iris[i]))],
      #'std': [st for i in range(len(df[i]))],
      'med': [median for i in range(len(iris[i]))],
    })

    df2.plot()
    plt.title('mean,data, and median of {}'.format(i))
    plt.xlabel(i)
    plt.show()
```
